```python
import cv2
from ultralytics import YOLO

# 1. Load a pre-trained YOLO model (nano version for fast performance)
model = YOLO("yolov8n.pt")

# 2. Open a connection to your webcam (0 is usually the default built-in camera)
cap = cv2.VideoCapture(0)

print("Starting object detection... Press 'q' to quit.")

while cap.isOpened():
    success, frame = cap.read()
    
    if not success:
        print("Error: Could not read frame from webcam.")
        break

    # 3. Run the model on the current webcam frame
    # stream=True optimizes memory usage for video feeds
    results = model(frame, stream=True)

    # 4. Draw the bounding boxes and labels onto the frame
    for result in results:
        annotated_frame = result.plot()

    # 5. Display the live video feed window
    cv2.imshow("Real-Time Object Detection", annotated_frame)

    # Break the loop if the 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 6. Clean up and close all windows
cap.release()
cv2.destroyAllWindows()
```
