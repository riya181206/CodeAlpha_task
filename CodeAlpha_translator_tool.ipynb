```python
from tkinter import *
from tkinter import ttk
from deep_translator import GoogleTranslator

root = Tk()
root.title("Language Translator")
root.geometry("1100x450")
# 1. Changed main window background to black
root.config(bg='black')

# Fetch a clean dictionary of all supported languages
LANGUAGES_DICT = GoogleTranslator().get_supported_languages(as_dict=True)
language_list = list(LANGUAGES_DICT.keys())

# 2. Centered Title Label at the top with matching black background
title_label = Label(root, text="TRANSLATOR TOOL", font="arial 20 bold", fg="white", bg="black")
title_label.pack(side=TOP, pady=20)

# --- Layout Elements (Adjusted coordinates to make space for the title) ---
Label(root, text='Input Text', font='arial 13 bold', fg='white', bg='black').place(x=100, y=90)
Input_text = Text(root, font='arial 10', height=5, wrap=WORD, padx=5, pady=5, width=50)
Input_text.place(x=100, y=130)

Label(root, text='Output Text', font='arial 13 bold', fg='white', bg='black').place(x=600, y=90)
Output_text = Text(root, font='arial 10', height=5, wrap=WORD, padx=5, pady=5, width=50)
Output_text.place(x=600, y=130)

# Dropdown menu configuration
dest_lang = ttk.Combobox(root, values=language_list, width=22, state="readonly")
dest_lang.place(x=100, y=260)
dest_lang.set('english')  

# --- Translation Logic ---
def Translate():
    try:
        user_text = Input_text.get(1.0, END).strip()
        target_language_name = dest_lang.get()

        if user_text and target_language_name in LANGUAGES_DICT:
            lang_code = LANGUAGES_DICT[target_language_name]
            translated_text = GoogleTranslator(source='auto', target=lang_code).translate(user_text)
            
            Output_text.delete(1.0, END)
            Output_text.insert(END, translated_text)
    except Exception as e:
        Output_text.delete(1.0, END)
        Output_text.insert(END, f"Error: Could not connect to translation server.")

# Button placement
trans_btn = Button(root, text='Translate', font='arial 12 bold', pady=5, command=Translate, bg='orange', activebackground='green')   
trans_btn.place(x=480, y=250)

root.lift()
root.attributes('-topmost', True)

root.mainloop()
```
